## Stack Implementation

The `Stack` class is an implementation of a stack data structure. The class comes with methods tailored for this PDA deliverable.

### Methods
- **`__init__(self)`**: Initializes a stack with `Z` as the first element and sets `top` to `Z`.
  
- **`push(self, element)`**: Adds an element to the top of the stack.

- **`pop(self)`**: Removes and returns the popped element from the stack.

- **`peek(self)`**: Returns the reversed elements of the stack as a string.

- **`is_empty(self)`**: Checks if the stack is empty and returns a boolean value indicating the result. Note that an empty list in this case is equivalent to `['Z']`


```python
class Stack:
  def __init__(self):
    self.stack = ['Z']
    self.top = 'Z'

  def push(self, element):
    self.stack.append(element)
    self.top = element

  def pop(self):
    if not self.is_empty():
      popped_item = self.stack.pop()
      if not self.is_empty():
        self.top = None
      else:
        self.top = self.stack[-1]
      return popped_item
  
  def peek(self):
    return ''.join(self.stack[::-1])
  
  def is_empty(self):
    return self.stack == ['Z']
```

```python
stack = Stack()

stack.push('A')
print(stack.peek())
print(stack.top)

stack.pop()
print(stack.peek())
print(stack.top)
```
